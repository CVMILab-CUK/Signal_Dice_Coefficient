```python
import os
import numpy as np
```

```python
path = "outputs"
with open(os.path.join(path,  "sample_result_mse.txt")) as f:
    lines = f.readlines()
```

```python
num_lines = []
for l in lines:
    num_lines.append(np.array(l[:-1].split()).astype(np.float32))
num_lines = np.array(num_lines)

mse_array, sdsc_array =num_lines[:, 0], num_lines[:, 1]

pearson_corr = np.corrcoef(mse_array, sdsc_array)[0, 1]
print("Pearson correlation:", pearson_corr)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.scatter(mse_array, sdsc_array, alpha=0.5)
plt.xlabel("MSE")
plt.ylabel("SDSC")
plt.title(f"Correlation: Pearson={pearson_corr:.3f}")
plt.grid(True)
plt.tight_layout()
plt.savefig("scatter_mse_sdsc.png", dpi=300)
plt.show()
```

```python
var_sdsc = np.var(sdsc_array)
std_sdsc = np.std(sdsc_array)

# IQR (?? 50%)
q25, q75 = np.percentile(sdsc_array, [25, 75])
iqr = q75 - q25

print("SDSC Std:", std_sdsc)
print("SDSC IQR:", iqr)
```

```python
mask = (mse_array > 1.25) & (mse_array < 1.75)
sds_sub = sdsc_array[mask]

var_sdsc = np.var(sds_sub)
std_sdsc = np.std(sds_sub)

# IQR (?? 50%)
q25, q75 = np.percentile(sdsc_array, [25, 75])
iqr = q75 - q25

print("SDSC Std:", std_sdsc)
print("SDSC IQR:", iqr)
```

```python
mask = (mse_array > 1.25) & (mse_array < 1.75)
sds_sub = sdsc_array[mask]

plt.hist(sds_sub, bins=20)
plt.title("SDSC distribution for MSE  1.5")
plt.xlabel("SDSC")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
```
